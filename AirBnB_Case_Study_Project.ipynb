```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"C:\Users\Owner\Documents\Week 3 Project\AB_NYC_2019.csv")
data
```

```python
host_rev_avg = data.groupby(['host_id','host_name'],as_index=False).sum(True).sort_values('reviews_per_month',ascending =False)
host_rev_avg[['host_id', 'host_name', 'reviews_per_month']].head()
#The busiest places are all in Manhattan and seem to be just hosts doing monthly rents for apts.
```

```python
# How many neighborhood groups are available and which shows up the most?
data.groupby('neighbourhood_group',as_index=False).count().sort_values('id',ascending=False)[['neighbourhood_group','id']]
#manhattan
```

```python
# Are private rooms the most popular in manhattan?
neighbor_data = data[["neighbourhood_group", "room_type",'reviews_per_month']].groupby(["neighbourhood_group", 'room_type'],as_index=False).sum(True).sort_values('reviews_per_month',ascending =False)
neighbor_data[neighbor_data["neighbourhood_group"] == "Manhattan"]
#Entire home/apt is the most popular
```

```python
# Which hosts are the busiest and based on their reviews?
host_rev_avg = data.groupby(['host_id','host_name'],as_index=False).sum(True).sort_values('number_of_reviews',ascending =False)
host_rev_avg[['host_id', 'host_name', 'number_of_reviews']].head()

#By reviews, Maya has the most
```

```python
#Which neighorhood group has the highest average price?
data.groupby('neighbourhood_group',as_index=False).mean(True).sort_values('price',ascending=False)[['neighbourhood_group','price']].round(2)
```

```python
# Which neighbor hood group has the highest total price?
data.groupby('neighbourhood_group',as_index=False).sum(True).sort_values('price',ascending=False)[['neighbourhood_group','price']]
```

```python
#Which top 5 hosts have the highest total price?
data.groupby(['host_id',"host_name"],as_index=False).sum(True).sort_values('price',ascending=False)[['host_id', 'host_name','price']].head()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
# review_bomb =data.groupby(['host_name','number_of_reviews','availability_365'],as_index=False).count()[['host_name','number_of_reviews',"availability_365"]]
# filterless =(review_bomb["number_of_reviews"] >=100) & (review_bomb["availability_365"] ==0)
# review_bomb[filterless]
my_data = data[(data["number_of_reviews"] >=100) & (data["availability_365"] ==0)]
my_data[['host_name','number_of_reviews',"availability_365"]]
```

```python
# What host has the highest total of prices and where are they located?
data.groupby(['host_id',"host_name",'neighbourhood_group'],as_index=False).sum(True).sort_values('price',ascending=False)[['host_id', 'host_name','price','neighbourhood_group']].head()
```

```python
# When did Danielle from Queens last receive a review?
dan_data = data[(data['host_name']== "Danielle")&(data['neighbourhood_group']=='Queens')&(data['last_review'].isnull() == False)].sort_values("last_review",ascending=False)
dan_data= dan_data.drop_duplicates("host_id")
dan_data[['host_id','host_name','neighbourhood_group','last_review']]
```
